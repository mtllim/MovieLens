## Section 1: Manipulating Data

```python
### Step 1: Import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# display plots inline
%matplotlib inline

# Links dataset
links = pd.read_csv("links.csv", skiprows=1, names=["movieId","imdbId","tmdbId"])
movies = pd.read_csv("movies.csv", skiprows=1, names=["movieId","title","genre1","genre2","genre3","genre4","genre5","genre6","genre7","genre8","genre9","genre10"])

#testing with a small subset
#movies_small = pd.read_csv("movies1.csv", skiprows=1, names=["movieId","title","genre1","genre2","genre3"])

ratings = pd.read_csv("ratings.csv",skiprows=1, names=["userId", "movieId", "rating", "timestamp"])
tags = pd.read_csv("tags.csv", skiprows=1, names=["userId","movieId","tag","timestamp"])

#print(ratings)
```

```python
# Step 2: Data Transformations

# Transpose from wide to long format
# https://pandas.pydata.org/pandas-docs/stable/reshaping.html#
# For each movie, there are many genres

movies_t = movies.melt(id_vars=['movieId', 'title'], var_name='g_num')
movies_t.columns = ['movieId', 'title','g_num','genre']
#print(movies_t)

# Summarise ratings file by movieId
# http://pandas.pydata.org/pandas-docs/stable/groupby.html
# For each movie, there are many ratings

ratings_summ = ratings.groupby('movieId').agg({'movieId': [np.min],
                                               'rating': [np.mean, np.sum]})
ratings_summ.columns = ['movieId', 'rating_mean', 'rating_sum']
#print(ratings_summ)

# Merge movies+genre file with movie+ratings file
# http://pandas.pydata.org/pandas-docs/stable/merging.html
# This file contains genres and ratings together

all_data = pd.merge(movies_t, ratings_summ, on='movieId')
print(all_data.head(n=5))

all_data1 = all_data.groupby('genre').agg({'rating_mean': [np.mean]})
all_data_stats = all_data.groupby('genre')['rating_mean'].agg(['count','mean']).reset_index()

#print(all_data_stats)
```

The sample data extract above shows that the first movie "Toy Story (1995)" has been labelled with 5 different genres and has a mean rating of 3.87 across all the users.

```python
print('The overall mean rating is',np.mean(all_data['rating_mean']))

plt.hist(all_data['rating_mean'],bins=10)
plt.title('Distribution of ratings across all genres')
plt.show()

plt.bar(all_data_stats['genre'], all_data_stats['count'])
plt.xticks(all_data_stats['genre'], all_data_stats['genre'], rotation=90)
plt.title('Number of ratings across genres')
plt.show()

plt.bar(all_data_stats['genre'], all_data_stats['mean'])
plt.xticks(all_data_stats['genre'], all_data_stats['genre'], rotation=90)
plt.title('Mean rating across genre')
plt.show()

all_data_stats1 = all_data_stats.sort_values('mean', ascending=False)
print(all_data_stats1)
#plt.bar(all_data_stats1['genre'], all_data_stats1['mean'])
#plt.xticks(all_data_stats1['genre'], all_data_stats1['genre'], rotation=90)
#plt.title('Mean rating across genre sorted descending')
plt.show()
```

The charts above show the descriptive statistics for the mean, across different genres. There is significant variation around the mean rating of 3.3.

The mean rating for each genre also varies significantly with different numbers of ratings for each genre.

The highest rated genres are Documentary, Film-Noir and War. The lowest rated genres and Action, IMAX and Horror.

## Section 2 Data Visualisation
### The "Wisdom of Crowds"

It has been said that the "wisdom of crowds" tend to be superior to those of individual experts. This traces back to an observation that Francis Galton pointed out that the crowd at a country fair accurately guessed the weight of an ox when their individual guesses were averaged, which was closer than any individual cattle expert.

Here we apply the same concept that the mean rating made by the collective wisdom of the movie audience reflects the true rating of the movie fairly well.

Here we define categories of similar genres to compare the ratings between categories and within genres. This is more informative than simply comparing all genres together at once as it's harder to see the differences.

The aim is to examine if any particular categories or individual genres are rated consistently higher or lower than others.
We will examine the mean of the ratings, the variance of the ratings and distribution of the ratings between genres.

```python
# Define categories
light_hearted_genres=['Animation','Children','Comedy','Musical','Romance']
exciting_genres=['Action','Adventure','Fantasy','IMAX','Sci-Fi']
dark_genres=['Film-Noir','Horror','Mystery','Thriller']
serious_genres=['Crime','Documentary','Drama','War','Western']


light_hearted=all_data.loc[all_data['genre'].isin(light_hearted_genres)]
exciting=all_data.loc[all_data['genre'].isin(exciting_genres)]
dark=all_data.loc[all_data['genre'].isin(dark_genres)]
serious=all_data.loc[all_data['genre'].isin(serious_genres)]

animation=all_data.loc[all_data['genre'].isin(['Animation'])]
children=all_data.loc[all_data['genre'].isin(['Children'])]
comedy=all_data.loc[all_data['genre'].isin(['Comedy'])]
musical=all_data.loc[all_data['genre'].isin(['Musical'])]
romance=all_data.loc[all_data['genre'].isin(['Romance'])]
action=all_data.loc[all_data['genre'].isin(['Action'])]
adventure=all_data.loc[all_data['genre'].isin(['Adventure'])]
fantasy=all_data.loc[all_data['genre'].isin(['Fantasy'])]
imax=all_data.loc[all_data['genre'].isin(['IMAX'])]
sci_fi=all_data.loc[all_data['genre'].isin(['Sci-Fi'])]
film_noir=all_data.loc[all_data['genre'].isin(['Film-Noir'])]
horror=all_data.loc[all_data['genre'].isin(['Horror'])]
mystery=all_data.loc[all_data['genre'].isin(['Mystery'])]
thriller=all_data.loc[all_data['genre'].isin(['Thriller'])]
crime=all_data.loc[all_data['genre'].isin(['Crime'])]
documentary=all_data.loc[all_data['genre'].isin(['Documentary'])]
drama=all_data.loc[all_data['genre'].isin(['Drama'])]
war=all_data.loc[all_data['genre'].isin(['War'])]
western=all_data.loc[all_data['genre'].isin(['Western'])]

```

### Section 2 Comparison across Categories of Genres




```python
bars = ('','light_hearted','exciting','dark','serious')
y_pos = np.arange(len(bars))
df1 = [light_hearted['rating_mean'], exciting['rating_mean'], dark['rating_mean'], serious['rating_mean']]

plt.boxplot(df1)
plt.xticks(y_pos, bars, rotation=90)
plt.title('Boxplot: Mean rating across category')
plt.show()

plt.violinplot(df1)
plt.xticks(y_pos, bars, rotation=90)
plt.title('Violinplot: Mean rating across category')
plt.show()
```

We see that the 'serious' genre has slightly higher mean rating and with less variability around this mean (seen in the violinplot).


## Section 3 ANOVA

The purpose of Analysis of Variance (ANOVA) test is to determine whether the means of different groups are the same, or if at least one is different. In this case, we are testing whether the mean of each category is significantly different, or whether the observed differences are simply due to random variability.

In the first graph below, the means of each group are the same, but the observed deviations are simply due to random variability.

In the second graph below, the means of each group are actually different, and we can conclude that the observed deviations are different enough to reject the null hypothesis that the means are the same.

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= "https://upload.wikimedia.org/wikipedia/commons/4/4b/Anova%2C_no_fit..png")
```

```python
Image(url= "https://upload.wikimedia.org/wikipedia/commons/8/80/ANOVA_fair_fit.jpg")
```

```python
# One-way ANOVA test for categories

from scipy import stats
f_value, p_value = stats.f_oneway(light_hearted['rating_mean'], exciting['rating_mean'], dark['rating_mean'], serious['rating_mean'])
print('f-value is',f_value, 'and p-value is',p_value)
```

We can perform an one-way ANOVA test to test whether any of the means of these 4 categories are significantly different from each other. As the p-value is <0.05, we can conclude that at least one of the categories have a significantly different mean.
This supports the observation that Serious movies are significantly higher rated.
